4. Customer Segmentation Using Clustering


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_profiles = df.groupby('cluster').mean()
print(cluster_profiles)
```

5. Data Visualization


```python
import seaborn as sns

sns.scatterplot(x='Recency', y='Income', hue='cluster', data=df, palette='Set1')
plt.title('Customer Segments by Recency and Income')
plt.show()
```

```python
cluster_profiles[['Income', 'MntTotal']].plot(kind='bar', figsize=(10, 6))
plt.title('Cluster Profiles')
plt.xlabel('Recency')
plt.ylabel('Customer_Days')
plt.show()
```
